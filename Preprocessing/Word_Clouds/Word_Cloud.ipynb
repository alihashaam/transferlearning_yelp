```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS

mpl.rcParams['figure.figsize']=(6.0,4.0)    #(6.0,4.0)
mpl.rcParams['font.size']=12                #10 
mpl.rcParams['savefig.dpi']=400             #72 
mpl.rcParams['figure.subplot.bottom']=.1 
```

```python
def word_cloud(df, Business):
    df.drop_duplicates(keep='first', inplace=True)
    df.dropna(axis=0, how='any', inplace=True)
    df_neg = df[df['stars'].isin([1,2])]
    df_pos = df[df['stars'].isin([4,5])]
    df_3 = df[df['stars'] ==3]
    
    wordcloud = WordCloud(background_color='white', stopwords=STOPWORDS, max_words=200, max_font_size=40, 
                              random_state=42).generate(str(df_pos['text']))
    fig = plt.figure(1,)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    fig.savefig(Business+"_Positive_Words.pdf")
    
    wordcloud_neg = WordCloud(background_color='white', stopwords=STOPWORDS, max_words=200, max_font_size=40, 
                              random_state=42).generate(str(df_neg['text']))
    fig_neg = plt.figure(2)
    plt.imshow(wordcloud_neg)
    plt.axis('off')
    plt.show()
    fig_neg.savefig(Business+"_Negative_Words.pdf")
    
    wordcloud_nutral = WordCloud(background_color='white', stopwords=STOPWORDS, max_words=200, max_font_size=40, 
                              random_state=42).generate(str(df_3['text']))
    fig_neutral = plt.figure(2)
    plt.imshow(wordcloud_nutral)
    plt.axis('off')
    plt.show()
    fig_neutral.savefig(Business+"_Neutral_Words.pdf")
```

```python
Business = 'Home_Services'
df =  pd.read_csv('F:\Books\Third Semester\DKE Project\dataset\Reviews_By_Business\Business_Home Services_reviews.csv')
word_cloud(df, Business)
```

```python
# a stacked bar plot with errorbars
import numpy as np
import matplotlib.pyplot as plt

ind = 1    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
p1 = []
bot = 0
lst = [0.78, 0.21, 0.79, 0.25]
lst_temp = sorted([x for x in lst if x!=max(lst)])
for x,y in zip(lst_temp, ["blue", "yellow", "green"]):
    p1.append(plt.bar(ind, x, bottom=bot, width=width, color=y))
    bot = bot+x

plt.ylabel('Accuracy')
plt.title('title')
plt.xticks([1])
plt.legend((p1[0], p1[1], p1[2]), ('1st', '2nd', '3rd'))

plt.show()
```
